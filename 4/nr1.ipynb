```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv", quotechar='"', sep=",")

variety = df["variety"]



figs, axs = plt.subplots(2, 3)

figs.set_figwidth(22)
figs.set_figheight(11)

data = df.groupby("variety")["sepal.length"].apply(list).tolist();
axs[0][0].hist(data, stacked=True, bins=25)
axs[0][0].legend(["setosa", "versicolor", "virgincia"])
axs[0][0].set_xlabel("sepal length (cm)")


data = df.groupby("variety")["petal.width"].apply(list).tolist();
axs[0][1].hist(data, stacked=True, bins=25)
axs[0][1].legend(["setosa", "versicolor", "virgincia"])
axs[0][1].set_xlabel("petal width (cm)")

data = df.groupby("variety")["petal.length"].apply(list).tolist();
axs[0][2].hist(data, stacked=True, bins=25)
axs[0][2].legend(["setosa", "versicolor", "virgincia"])
axs[0][2].set_xlabel("petal length (cm)")


data = df.groupby("variety")["sepal.width"].apply(list).tolist();
axs[1][0].hist(data, stacked=True, bins=25)
axs[1][0].legend(["setosa", "versicolor", "virgincia"])
axs[1][0].set_xlabel("sepal width (cm)")

for ax in axs.flat:
    if not ax.has_data():
        ax.set_visible(False)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv", quotechar='"', sep=",")


figs, axs = plt.subplots(1,2)
figs.set_figwidth(15)

axs[0].scatter(df["sepal.length"], df["sepal.width"], marker="+")
axs[0].set_xlabel("sepal length (cm)")
axs[0].set_ylabel("sepal with (cm)")

axs[1].scatter(df["petal.length"], df["petal.width"], marker=".", color=[2/255,101/255,2/255])
axs[1].set_xlabel("petal length (cm)")
axs[1].set_ylabel("petal width (cm)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics

df = pd.read_csv("iris.csv", quotechar='"', sep=",")

stdPetalLength = statistics.stdev(df["petal.length"])
stdPetalWidth = statistics.stdev(df["petal.width"])

covariance = np.cov(df["petal.length"], df["petal.width"])[0, 1]

korrelationsEffiziente = covariance / (stdPetalLength * stdPetalWidth)

print(korrelationsEffiziente)

stdPetalLength = statistics.stdev(df["sepal.length"])
stdPetalWidth = statistics.stdev(df["sepal.width"])

covariance = np.cov(df["sepal.length"], df["sepal.width"])[0, 1]

korrelationsEffiziente = covariance / (stdPetalLength * stdPetalWidth)

print(korrelationsEffiziente)
```
